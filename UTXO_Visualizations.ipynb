# Data Import

```
from google.colab import files
import pandas as pd
import decimal
from datetime import date, timedelta, datetime
import plotly.graph_objects as go
from plotly.offline import iplot
import plotly.express as px

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
  
result = pd.read_csv('result0911Task123.csv')
```

# Task 1

```
result['net_new'] = result['newborn'] - result['dead']
result['UTXO_Cum'] = result['net_new'].cumsum()

# Create traces
fig1 = go.Figure()
fig1.add_trace(go.Scatter(x=result.date, y=result.newborn,
                    mode='lines',
                    name='UTXO Newborn'))
fig1.add_trace(go.Scatter(x=result.date, y=result.dead,
                    mode='lines+markers',
                    name='UTXO Dead'))
fig1.add_trace(go.Scatter(x=result.date, y=result.net_new,
                    mode='markers', name='UTXO Netnew'))
fig1.update_layout(title='Daily Newborn and Dead UTXO',
                   xaxis_title='Date',
                   yaxis_title='UTXO')
fig1.update_yaxes(type="log")

# Add range slider
fig1.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

iplot(fig1)
```

```
# Create traces
fig11 = go.Figure()
fig11.add_trace(go.Scatter(x=result.date, y=result.UTXO_Cum,
                    mode='markers', name='UTXO Cum'))
fig11.update_layout(title='The Accumulated UTXO',
                   xaxis_title='Date',
                   yaxis_title='UTXO')
# Add range slider
fig11.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
iplot(fig11)
```

```
result['dead_cum'] = result['dead'].cumsum()
result['lifelength_cum'] = (result['dead'].mul(result['WALE'])).cumsum()
result['WALE_cum'] = result['lifelength_cum']/result['dead_cum'] 
result
```

```
fig2 = px.line(result, x = 'date', y = 'WALE_cum', title = 'The Weighted Average of Life Expectancy by UTXO Value')
# Add range slider
fig2.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)
iplot(fig2)
```

```
categories = [-9, -7, -5, -3, -1, 1, 3, 5, 7, 9]
for i in range(0, len(categories)):
  working_index = int(categories[i])
  result[str(working_index)] = result[str(working_index)].cumsum()
```

```
import plotly.offline as py     
import plotly.graph_objects as go

trace0 = go.Scatter(x = result["date"], y = result["-9"], name = '< 1d')
trace1 = go.Scatter(x = result["date"], y = result["-7"], name = '1d ~ 1m')
trace2 = go.Scatter(x = result["date"], y = result["-5"], name = '1m ~ 1q')
trace3 = go.Scatter(x = result["date"], y = result["-3"], name = '1q ~ 6m')
trace4 = go.Scatter(x = result["date"], y = result["-1"], name = '6m ~ 1y')
trace5 = go.Scatter(x = result["date"], y = result["1"], name = '1y ~ 2y')
trace6 = go.Scatter(x = result["date"], y = result["3"], name = '2y ~ 3y')
trace7 = go.Scatter(x = result["date"], y = result["5"], name = '3y ~ 4y')
trace8 = go.Scatter(x = result["date"], y = result["7"], name = '4y ~ 5y')
trace9 = go.Scatter(x = result["date"], y = result["9"], name = '> 5y')

layout = go.Layout(title="life Expectancy Distribution", xaxis = dict(title="date"), yaxis = dict(title ="Life Expectancy Distribution") ) 
data = [trace0, trace1, trace2, trace3, trace4, trace5, trace6, trace7, trace8, trace9]
fig = go.Figure(layout = layout, data = data) 
fig.update_yaxes(type="log")

# Add range slider
fig.update_layout(
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label="1m",
                     step="month",
                     stepmode="backward"),
                dict(count=6,
                     label="6m",
                     step="month",
                     stepmode="backward"),
                dict(count=1,
                     label="YTD",
                     step="year",
                     stepmode="todate"),
                dict(count=1,
                     label="1y",
                     step="year",
                     stepmode="backward"),
                dict(step="all")
            ])
        ),
        rangeslider=dict(
            visible=True
        ),
        type="date"
    )
)

py.iplot(fig)
```
